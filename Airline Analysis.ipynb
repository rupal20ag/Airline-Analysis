```python
#how the number of tickets booked and total amount earned changed with the time.
tickets=pd.read_sql_query("""select * 
from tickets as t
join bookings as b
on t.book_ref=b.book_ref""",connection)

tickets['book_date']=pd.to_datetime(tickets['book_date'])
tickets.dtypes
tickets['date']=tickets['book_date'].dt.date
tickets
x=tickets.groupby(['date']).agg({'ticket_no':'count'})
plt.figure(figsize=(18,6))
plt.plot(x.index,x.values,marker='^')
plt.xlabel('Date',fontsize=20)
plt.ylabel('Tickets_booked',fontsize=20)
plt.title('Total Tickets booked by Date',fontsize=25)
plt.grid('b')
plt.show()
```

Analysing Occupancy Rate

```python
#for each aircraft, calculate the total revenue per year and the average revenue per ticket.
pd.read_sql_query("""
with cte as
(select sum(tf.amount) as total_rev,count(tf.ticket_no) as total_tickets,f.aircraft_code
from ticket_flights tf
join flights f
on tf.flight_id=f.flight_id
group by 3)
select total_rev,avg(total_rev/total_tickets) as avg_rev_per_ticket,total_tickets,aircraft_code
from cte
group by 4
order by 1
""",connection)
```

```python
#Calculate the average occupancy per aircraft
Occupancy_rate=pd.read_sql_query("""WITH cte AS (
    SELECT 
        aircraft_code,
        f.flight_id,
        COUNT(bp.seat_no) AS seats
    FROM 
        boarding_passes bp
    JOIN 
        flights f ON bp.flight_id = f.flight_id
    GROUP BY 
        1, 2
),
total_seats AS (
    SELECT 
        aircraft_code,
        COUNT(seat_no) AS total_seats
    FROM 
        seats
    GROUP BY 
        aircraft_code
)
SELECT 
    cte.aircraft_code,
    total_seats.total_seats AS total_seats,
    cast(AVG(cte.seats) as integer) AS booked_seats,
    AVG(cte.seats)*100 / total_seats.total_seats AS occupancy_rate
FROM 
    cte
JOIN 
    total_seats ON cte.aircraft_code = total_seats.aircraft_code
GROUP BY 
    1, 2
""",connection)
Occupancy_rate
```

```python
#Calculate the average occupancy per aircraft
pd.read_sql_query("""SELECT 
        aircraft_code,
        count(distinct f.flight_id)
    FROM 
        boarding_passes bp
    JOIN 
        flights f ON bp.flight_id = f.flight_id
    GROUP BY 
        1""",connection)
```

```python
#calculate by how much total annual turnover could increase by giving all aircraft a 10% of higher occupancy rate.
```

```python
Occupancy_rate['Increased Occ rate']=Occupancy_rate['occupancy_rate']*1.10
Occupancy_rate
```

```python
pd.set_option("display.float_format",str)
total_revenue=pd.read_sql_query("""select aircraft_code,sum(amount) as total_rev
from ticket_flights tf
join flights f
on tf.flight_id=f.flight_id
group by 1""",connection)
total_revenue
Occupancy_rate['Inc Total Annual Turnover']=(total_revenue['total_rev']/Occupancy_rate['occupancy_rate'])*Occupancy_rate['Increased Occ rate']
Occupancy_rate
def format_turnover(turnover):
    units = ['', 'K', 'M', 'B', 'T']
    for unit in units[3:]:
        if abs(turnover) >= 1000:
            turnover /= 1000.0
        else:
            break
    return f'{turnover:.2f} {unit}'
    
Occupancy_rate['Inc Total Annual Turnover'] = Occupancy_rate['Inc Total Annual Turnover'].map(format_turnover)
Occupancy_rate
```
